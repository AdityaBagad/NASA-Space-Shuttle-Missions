```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
# Import the dataset
budget_data = pd.read_csv('NASA-Budget.csv')
```

```python
budget_data.head()
```

```python
# Create unique lists for each columns
year = list(budget_data['Year'].astype('int64'))
fed_budget = list(budget_data['FederalBudget'].astype('float'))
nominal_dollars = list(budget_data['NominalDollars'].astype('int64'))
constant_dollars = list(budget_data['ConstantDollars'].astype('int64'))
```

```python
# Scatter plot to show Year vs Dollars Expenditure
plt.figure(figsize=(15,5))

plt.title('NASA Budget',size=20)
plt.xlabel('Year', size=15)
plt.ylabel('Budget', size=15)

plt.plot(year, nominal_dollars, label='Nominal Dollars')
plt.plot(year, constant_dollars, label='Constant Dollars')

plt.xticks(np.arange(min(year), max(year)+1, 3))
plt.legend()

plt.show()
```

```python
# Plot the Federal Budget Year wise
plt.figure(figsize=(15,10))

plt.bar(year, fed_budget, label='Federal Budget')

plt.title('NASA Budget as a Percentage of Federal Budget',size=20)
plt.xlabel('Year', size=15)
plt.ylabel('Percentage of Federal Budget', size=15)

plt.xticks(np.arange(min(year), max(year)+1, 3))
plt.yticks(np.arange(0, max(fed_budget), 0.2))

plt.legend()

plt.show()
```
