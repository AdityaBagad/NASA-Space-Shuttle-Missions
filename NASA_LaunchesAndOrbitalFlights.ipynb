```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
# Import the dataset
flights_data = pd.read_csv('LaunchesAndOrbitalFlights.csv')
```

```python
flights_data.head()
```

```python
# Create unique lists for each columns
year = list(flights_data['Year'])
shuttle = list(flights_data['Shuttle'])
landingsite = list(flights_data['LandingSite'])
```

```python
# Scatter plot to show Year vs Shuttle
plt.figure(figsize=(15,5))

plt.title('NASA Space Shuttle Launches',size=20)
plt.xlabel('Year', size=15)
plt.ylabel('Shuttle Name', size=15)

plt.scatter(year, shuttle)

plt.xticks(np.arange(min(year), max(year)+1, 2))

plt.show()
```

```python
sns.set(font_scale=1.3)

sns.set_style('whitegrid')

scatterData = pd.DataFrame(data=[flights_data['Year'], flights_data['Shuttle']]).T

plt.figure(figsize=(15,5))

sns.scatterplot(x="Year", y="Shuttle",data=scatterData, hue='Shuttle')

plt.xticks(np.arange(min(year), max(year)+1, 2))
plt.title('NASA Space Shuttle Launches')

plt.show()
```
